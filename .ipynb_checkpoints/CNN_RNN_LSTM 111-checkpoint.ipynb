## **3.1. define functions**

Defined 8 functions, each are:
1.  get_all_str(): to get all of strings from dataframe column.
2. get_str(): get string from list
3. get_word(): to get words from text, using RegexpTokenizer
4. stopword_list(): to add stopwords to nltp stopword list
5. remove_stopword(): to remove stopwords from list
6. Freq_df(): to get dataframe from cleanwordlist
7. lemmatization(): to lemmatize words
and also visualization function:
8. Word_Cloud(): plot word cloud of words

```python
# function to get all of strings from dataframe column, and used lower function here.
def get_all_str(df):
    sentence = ''
    for i in range(len(df)):
        sentence += df['Message'][i]
    sentence = sentence.lower()
    return sentence

def get_str(lst):
    sentence = ''
    for char in lst:
        sentence += char+' '
    sentence = sentence.lower()
    return sentence

# function to get words from text(string). used RegexpTokenizer
def get_word(text):
    result = nltk.RegexpTokenizer(r'\w+').tokenize(text.lower())
#     result = result.lower()
#     result = nltk.word_tokenize(text)
    return result

# function to add stopwords to nltp stopword list.
def stopword_list(stop):
    lst = stopwords.words('english')
    for stopword in stop:
        lst.append(stopword)
    return lst

# function to remove stopwords from list.
def remove_stopword(stopwords, lst):
    stoplist = stopword_list(stopwords)
    txt = ''
    for idx in range(len(lst)):
        txt += lst[idx]
        txt += '\n'
    cleanwordlist = [word for word in txt.split() if word not in stoplist]
    return cleanwordlist

# function to get dataframe from cleanwordlist.
def Freq_df(cleanwordlist):
    Freq_dist_nltk = nltk.FreqDist(cleanwordlist)
    df_freq = pd.DataFrame.from_dict(Freq_dist_nltk, orient='index')
    df_freq.columns = ['Frequency']
    df_freq.index.name = 'Term'
    df_freq = df_freq.sort_values(by=['Frequency'],ascending=False)
    df_freq = df_freq.reset_index()
    return df_freq

# function to lemmatize words
def lemmatization(words):
    lemm = WordNetLemmatizer()
    tokens = [lemm.lemmatize(word) for word in words]
    return tokens
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dense, Dropout, BatchNormalization, GlobalAveragePooling1D


X = df['Message']
y = df['Category']

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state = 8888)
max_len = max(max(train_X.apply(len).values),max(test_X.apply(len).values))

def fit_tokenizer(text, oov_token):
    tokenizer = Tokenizer(oov_token = oov_token)
    tokenizer.fit_on_texts(text)
    return tokenizer

# for sequence, padding
def seq_padding(sentences, tokenizer, padding, truncating, maxlen):
    sequences = tokenizer.texts_to_sequences(sentences)
    pad_trunc_sequences = pad_sequences(sequences, padding = padding, maxlen = maxlen, truncating=padding)
    return pad_trunc_sequences

tokenizer = fit_tokenizer(train_X, "<OOV>")

word_index = tokenizer.word_index
VOCAB_SIZE = len(word_index)

train_X = seq_padding(train_X,tokenizer, 'post', 'post',max_len)
test_X = seq_padding(test_X,tokenizer, 'post', 'post', max_len)

def callback(MATRIX, PATIENCE):

    callbacks = [EarlyStopping(monitor=MATRIX,
                                patience=PATIENCE,
                                restore_best_weights=True),
                  ModelCheckpoint('model.weights.h5',
                                  monitor=MATRIX,
                                  save_best_only=True,
                                  save_weights_only=True)]
    return callbacks
```

# **CNN_Model**

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization

def CNN_model(input_shape, num_classes):
    model = Sequential()

    # Adjusted Convolutional Block for non-square input (e.g., 70x13)
    model.add(Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(70, 13, 1)))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))

    # Additional layers remain the same
    model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))

    model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))

    # Flatten and Dense Layers
    model.add(Flatten())
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))

    # Output Layer
    model.add(Dense(num_classes, activation='softmax'))

    # Compile the model
    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    return model


input_shape = (max_len, 100) # Example input shape - adjust as needed

model = CNN_model(input_shape, 100) # Pass the correct input shape

model.summary()

# Add the following line before calling model.summary()
# model.build(input_shape=(None, max_len)) # No longer needed if input_shape is defined correctly

# Show the model architecture with a smaller size
tf.keras.utils.plot_model(
    model,
    show_shapes=True,
    show_layer_names=True,
    dpi=60  # Adjust DPI for smaller image size
)
```

# **RNN_Model**

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, Dropout, BatchNormalization, Bidirectional

def RNN_model(vocab_size, embedding_dim, maxlen):

    model = Sequential()

    # Embedding layer
    model.add(Embedding(vocab_size + 1, embedding_dim, input_length=maxlen))

    # First RNN Layer
    model.add(Bidirectional(SimpleRNN(128, return_sequences=True, kernel_regularizer=tf.keras.regularizers.l2(0.001))))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))

    # Second RNN Layer
    model.add(Bidirectional(SimpleRNN(64, return_sequences=False, kernel_regularizer=tf.keras.regularizers.l2(0.001))))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))

    # Dense Layers
    model.add(Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)))
    model.add(Dropout(0.5))
    model.add(Dense(32, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)))
    model.add(Dropout(0.5))

    # Output Layer
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model
    model.compile(loss='binary_crossentropy',
                  optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                  metrics=['accuracy'])

    return model

model = RNN_model(VOCAB_SIZE, 100, max_len)

model.summary()

# Add the following line before calling model.summary()
model.build(input_shape=(None, max_len))

# Show the model architecture with a smaller size
tf.keras.utils.plot_model(
    model,
    show_shapes=True,
    show_layer_names=True,
    dpi=60  # Adjust DPI for smaller image size
)
```

# **LSTM_Model**

```python
def LSTM_model(vocab_size, embedding_dim, maxlen):

    model = Sequential()

    # Embedding layer
    model.add(Embedding(vocab_size + 1, embedding_dim, input_length=maxlen))

    # First Bidirectional LSTM layer
    model.add(Bidirectional(LSTM(128, return_sequences=True, kernel_regularizer=tf.keras.regularizers.l2(0.001))))
    model.add(BatchNormalization())  # Batch Normalization
    model.add(Dropout(0.5))  # Dropout for regularization

    # Second Bidirectional LSTM layer
    model.add(Bidirectional(LSTM(64, return_sequences=True, kernel_regularizer=tf.keras.regularizers.l2(0.001)))) # Changed return_sequences to True
    model.add(BatchNormalization())  # Batch Normalization
    model.add(Dropout(0.5))  # Dropout for regularization

    # Global Average Pooling layer
    model.add(GlobalAveragePooling1D())

    # Dense layers
    model.add(Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)))
    model.add(Dropout(0.5))  # Dropout for regularization
    model.add(Dense(6, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)))
    model.add(Dropout(0.5))  # Dropout for regularization

    # Output layer
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model
    model.compile(loss='binary_crossentropy',
                  optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                  metrics=['accuracy'])

    return model


model = LSTM_model(VOCAB_SIZE, 100, max_len)

model.summary()

# Add the following line before calling model.summary()
model.build(input_shape=(None, max_len))

# Show the model architecture with a smaller size
tf.keras.utils.plot_model(
    model,
    show_shapes=True,
    show_layer_names=True,
    dpi=60  # Adjust DPI for smaller image size
)
```

```python
# Evaluate each model
def evaluate_model(model, test_X, test_y):
    test_prediction_label = (model.predict(test_X) >= 0.5).astype("int32")
    test_predict_proba = model.predict(test_X, verbose=0)

    accuracy = accuracy_score(test_y, test_prediction_label)
    auc = roc_auc_score(test_y, test_predict_proba)
    recall = recall_score(test_y, test_prediction_label)
    f1 = f1_score(test_y, test_prediction_label)

    return accuracy, auc, recall, f1

# Store metrics for each model
model_names = ['CNN', 'RNN', 'LSTM']
accuracy_list = []
auc_list = []
recall_list = []
f1_list = []

# Evaluate and store results for each model
for model in [cnn_model, rnn_model, lstm_model]:
    accuracy, auc, recall, f1 = evaluate_model(model, test_X, test_y)
    accuracy_list.append(accuracy)
    auc_list.append(auc)
    recall_list.append(recall)
    f1_list.append(f1)

# Print the results
for i, model_name in enumerate(model_names):
    print(f"{model_name} - Accuracy: {accuracy_list[i]:.4f}, AUC: {auc_list[i]:.4f}, Recall: {recall_list[i]:.4f}, F1 Score: {f1_list[i]:.4f}")
```
